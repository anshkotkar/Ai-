# Analysisng the Rateing



```python
# storeing country name inside the variables
# country_name is a variable

country_name = df_final.Country.value_counts().index
print(country_name)
```

```python
# storeing country values inside the variables

country_value = df_final.Country.value_counts().values
print(country_value)
```
